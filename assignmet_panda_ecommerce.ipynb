```python
new_df=order.merge(product, on = 'product_id')
new_df['order_date']= pd.to_datetime(new_df.order_date)
new_df['year']= new_df.order_date.dt.year
new_df['month']= new_df.order_date.dt.month
new_df['total_sale'] = new_df.quantity * new_df.price
year_grp=new_df.groupby('year')
new_df
```

```python
data =customer.merge(new_df, on = 'customer_id')
data
```

```python
def agegroup(x):
    if x >= 50:
        return 'sr_citizen'
    elif x >= 30:
        return 'Adult'
    elif x >= 20:
        return  'Young'
    elif x >= 15:
        return 'Gen_z'
```

```python
idss =data.groupby('customer_id')
id=data['customer_id'].value_counts().index
value = data['customer_id'].value_counts().values
new_data = pd.DataFrame(id)
new_data['repeat_order'] = value
new_df1=new_data.merge(customer, on= "customer_id")
new_df1['age_category'] = new_df1['age'].apply(agegroup)
new_df1
```

```python
cate = []
price= []
for ci in id:
    ne = idss.get_group(ci).category
    pri = idss.get_group(ci).price.sum()
    cate.append(ne.mode().sum())
    price.append(pri)
new_df1['category']=cate
new_df1['total_purchase'] = price
new_df1
```

```python
category_re =new_df1['category'].value_counts().index
cate_num_repeat=new_df1['category'].value_counts().values
category_repeat= pd.DataFrame(category_re)
category_repeat['category_repeat'] = cate_num_repeat
```

```python
name_pro=data['product_name'].value_counts().index
value_pro=data['product_name'].value_counts().values
repeat_product=pd.DataFrame(name_pro)
repeat_product['repeat_product'] = value_pro
new_df2=repeat_product.merge(product, on = 'product_name')
```

```python
new_df3=new_df2.merge(category_repeat, on= 'category')
```

```python
new_df3 = new_df3[['product_id', 'price', 'product_name', 'repeat_product','category','category_repeat']]
```
