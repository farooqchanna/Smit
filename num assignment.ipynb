```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('InvestmentDataset.csv')
```

```python
per=np.array(data['Profit Percentage']).astype(str)
```

```python
per = np.char.rstrip(per, "%").astype(int)
year = np.array(data['Year'])
A_invest = np.array(data['Additional Investment'])
total = np.array(data['Total Amount'])
```

```python
total = total + A_invest
```

```python
pl = []
initial = 5000
for number in total:
    Total = number - initial
    initial = number
    pl.append(Total)
pl = np.array(pl)
g_m = np.power(pl.prod(),1/len(pl))
```

```python
gm_total = np.power(total.prod(),1/len(total))
```

```python
total_P= total[-1]-5000
```
